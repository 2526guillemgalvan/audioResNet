```python
import os

AUDIO_DIR = "ESC-50-master/audio"
IMAGE_DIR = "dataset_jpg"

def check_dataset(audio_dir, image_dir):
    # Listar archivos
    audio_files = sorted([
        f for f in os.listdir(audio_dir)
        if f.endswith(".wav")
    ])

    image_files = sorted([
        f for f in os.listdir(image_dir)
        if f.endswith(".jpg")
    ])

    # Conjuntos base (sin extensi√≥n)
    audio_names = {os.path.splitext(f)[0] for f in audio_files}
    image_names = {os.path.splitext(f)[0] for f in image_files}

    # Comparaciones
    missing_images = audio_names - image_names
    extra_images = image_names - audio_names

    print("üìä RESUMEN")
    print(f"Audios (.wav): {len(audio_files)}")
    print(f"Im√°genes (.jpg): {len(image_files)}")

    if not missing_images and not extra_images:
        print("‚úÖ Todas las conversiones son correctas")
    else:
        print("‚ùå Problemas encontrados:")

        if missing_images:
            print(f"  - Faltan {len(missing_images)} im√°genes para estos audios:")
            for name in sorted(missing_images):
                print(f"    ‚ùå {name}.jpg")

        if extra_images:
            print(f"  - Hay {len(extra_images)} im√°genes sin audio:")
            for name in sorted(extra_images):
                print(f"    ‚ö†Ô∏è {name}.jpg")


if __name__ == "__main__":
    check_dataset(AUDIO_DIR, IMAGE_DIR)
```
